# Time series barycenters

`tslearn` provides a `barycenter` package that offers ways to compute time series barycenters.

```python
%pylab inline
import sys
sys.path.append("../../../")

from tslearn.barycenters import EuclideanBarycenter
from tslearn.generators import random_walk_blobs

n_ts, sz, d = 20, 128, 1

numpy.random.seed(0)
X, y = random_walk_blobs(n_ts_per_blob=n_ts, sz=sz, d=d, noise_level=2., n_blobs=1)

for ts in X:
    pylab.plot(numpy.arange(sz), ts, "k-", alpha=.2)
pylab.plot(numpy.arange(sz), EuclideanBarycenter().fit(X), "r-", linewidth=2);
```
