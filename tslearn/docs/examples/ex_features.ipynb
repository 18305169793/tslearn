# Time series transformers

`tslearn` provides a `piecewise` package that contains Piecewise Aggregate Approximation (PAA) and Symbolic Aggregate approXimation (SAX).

```python
%pylab inline

from tslearn.generators import random_walks
from tslearn.preprocessing import TimeSeriesScalerMeanVariance
from tslearn.piecewise import PiecewiseAggregateApproximation
from tslearn.piecewise import SymbolicAggregateApproximation, OneD_SymbolicAggregateApproximation

numpy.random.seed(0)
n_ts, sz, d = 2, 100, 1
dataset = random_walks(n_ts=n_ts, sz=sz, d=d)
scaler = TimeSeriesScalerMeanVariance(mu=0., std=1.)  # Rescale time series
dataset = scaler.fit_transform(dataset)

pylab.plot(numpy.arange(sz), dataset[0, :, 0], "g-")
pylab.plot(numpy.arange(sz), dataset[1, :, 0], "b-");
```

## Piecewise Aggregate Approximation (PAA)

Let us first visualize how PAA transforms our time series:

```python
n_paa_segments = 10
paa = PiecewiseAggregateApproximation(n_segments=n_paa_segments)
paa_dataset = paa.fit_transform(dataset)

paa_dataset_inv = paa.inverse_transform(paa_dataset)

pylab.figure(figsize=(12, 5))
pylab.subplot(1, 2, 1)
pylab.plot(numpy.arange(sz), dataset[0, :, 0], "g-", alpha=0.4)
pylab.plot(numpy.arange(sz), paa_dataset_inv[0], "g-")

pylab.subplot(1, 2, 2)
pylab.plot(numpy.arange(sz), dataset[1, :, 0], "b-", alpha=0.4)
pylab.plot(numpy.arange(sz), paa_dataset_inv[1], "b-");
```

We can now compute PAA distance by two means:

```python
print(paa.distance_paa(paa_dataset[0], paa_dataset[1]))
print(paa.distance(dataset[0], dataset[1]))
```

In practice, if you already have your PAA transformed time series, you should use `distance_paa` for the sake of efficiency, otherwise, `distance` can compute the same quantity starting from raw time series. Finally, we can check that these quantities lower bound the Euclidean Distance:

```python
print(numpy.linalg.norm(dataset[0] - dataset[1]))
```

## Symbolic Aggregate approXimation (SAX)

Let us now turn our focus on SAX transformation for time series:

```python
n_sax_symbols = 3
sax = SymbolicAggregateApproximation(n_segments=n_paa_segments, alphabet_size_avg=n_sax_symbols)
sax_dataset = sax.fit_transform(dataset)

pylab.plot(numpy.arange(sz), dataset[0, :, 0], "g-")
pylab.plot(numpy.arange(sz), dataset[1, :, 0], "b-")
for bp in sax.breakpoints_avg_:
    pylab.axhline(y=bp, linestyle="dashed", color="k", alpha=0.4)
```

```python
sax_dataset_inv = sax.inverse_transform(sax_dataset)

pylab.figure(figsize=(12, 5))
pylab.subplot(1, 2, 1)
pylab.plot(numpy.arange(sz), dataset[0, :, 0], "g-", alpha=0.4)
pylab.plot(numpy.arange(sz), sax_dataset_inv[0], "g-")

pylab.subplot(1, 2, 2)
pylab.plot(numpy.arange(sz), dataset[1, :, 0], "b-", alpha=0.4)
pylab.plot(numpy.arange(sz), sax_dataset_inv[1], "b-");
```

Finally, just as with PAA, we can compute SAX distance in two ways (and observe that it lower bounds the PAA distance):

```python
print(sax.distance_sax(sax_dataset[0], sax_dataset[1]))
print(sax.distance(dataset[0], dataset[1]))
```

## 1d-Symbolic Aggregate approXimation (1d-SAX)

One can also use 1d-SAX in which both average values and slopes are encoded for each segment:

```python
n_sax_symbols_avg = 4
n_sax_symbols_slope = 4

one_d_sax = OneD_SymbolicAggregateApproximation(n_segments=n_paa_segments, alphabet_size_avg=n_sax_symbols_avg,
                                                alphabet_size_slope=n_sax_symbols_slope)
one_d_sax_dataset = one_d_sax.fit_transform(dataset)
one_d_sax_dataset_inv = one_d_sax.inverse_transform(one_d_sax_dataset)

pylab.figure(figsize=(12, 5))
pylab.subplot(1, 2, 1)
pylab.plot(numpy.arange(sz), dataset[0, :, 0], "g-", alpha=0.4)
pylab.plot(numpy.arange(sz), one_d_sax_dataset_inv[0], "g-")

pylab.subplot(1, 2, 2)
pylab.plot(numpy.arange(sz), dataset[1, :, 0], "b-", alpha=0.4)
pylab.plot(numpy.arange(sz), one_d_sax_dataset_inv[1], "b-");
```

```python
print(one_d_sax.distance_1d_sax(one_d_sax_dataset[0], one_d_sax_dataset[1]))
print(one_d_sax.distance(dataset[0], dataset[1]))
```
